```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import os

%matplotlib inline
```

```python
files = glob.glob("*.dat")
total_rows = 0
```

```python
datasets = {}
for file in files:
    filename = os.path.basename(file).replace(".dat","")
    print(filename)
    df = pd.read_csv(file,delimiter="\s+", names=["time","elevation"])
    datasets[filename] = df
    df = df[np.logical_and(df.time < 60*30, df.time > 0)]
    print(len(df.elevation))
    total_rows += len(df.elevation)
    plt.plot(df.time,df.elevation)
    plt.show()
print(total_rows)
```
